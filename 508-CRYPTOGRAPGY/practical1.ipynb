```python
#1)program to implement password salting and hashing to create super passwords

import hashlib
import os

def generate_salt():
    return os.urandom(16)  # Generates a random 16-byte (128-bit) salt

def hash_password(password, salt):
    # Combine the password and salt and hash them using SHA-256
    hashed_password = hashlib.sha256(password.encode() + salt).hexdigest()
    return hashed_password

def create_super_password():
    # Get user input for password
    password = input("Enter your password: ")

    # Generate a random salt
    salt = generate_salt()

    # Hash the password with the generated salt
    hashed_password = hash_password(password, salt)

    # Return the hashed password and the salt
    return hashed_password, salt

super_password, salt = create_super_password()
print(f"Hashed Password: {super_password}")
print(f"Salt: {salt}")
```

```python
#3)program to demostrate cryptanalysis(e.g breaking caesar or  Vigenère Cipher) 

def caesar_decrypt(ciphertext, shift):
    plaintext = ""
    for char in ciphertext:
        if char.isalpha():
            ascii_offset = ord('a') if char.islower() else ord('A')
            decrypted_char = chr((ord(char) - ascii_offset - shift) % 26 + ascii_offset)
            plaintext += decrypted_char
        else:
            plaintext += char
    return plaintext

def caesar_break(ciphertext):
    for shift in range(16):
        decrypted_message = caesar_decrypt(ciphertext, shift)
        print(f"Shift {shift}: {decrypted_message}")

# Example usage:
ciphertext = "Khoor zruog, wklv lv d vhfuhw phvvdjh."
print(f"Encrypted message: {ciphertext}")
print("\nAttempting to break Caesar Cipher:")
caesar_break(ciphertext)



```

```python
#4)program to implement AES algorithm for file encryption and decryption

from cryptography.hazmat.primitives import padding
from cryptography.hazmat.primitives.ciphers import Cipher, algorithms, modes
from cryptography.hazmat.backends import default_backend

def pad(data):
    padder = padding.PKCS7(128).padder()
    return padder.update(data) + padder.finalize()

def unpad(data):
    unpadder = padding.PKCS7(128).unpadder()
    return unpadder.update(data) + unpadder.finalize()

def encrypt_file(input_file, output_file, key):
    with open(input_file, 'rb') as file:
        plaintext = file.read()

    plaintext = pad(plaintext)

    iv = b'\x00' * 16  # You should generate a random IV for each encryption

    cipher = Cipher(algorithms.AES(key), modes.CFB(iv), backend=default_backend())
    encryptor = cipher.encryptor()
    ciphertext = encryptor.update(plaintext) + encryptor.finalize()

    with open(output_file, 'wb') as file:
        file.write(ciphertext)

def decrypt_file(input_file, output_file, key):
    with open(input_file, 'rb') as file:
        ciphertext = file.read()

    iv = b'\x00' * 16  # You should use the same IV used for encryption

    cipher = Cipher(algorithms.AES(key), modes.CFB(iv), backend=default_backend())
    decryptor = cipher.decryptor()
    plaintext = decryptor.update(ciphertext) + decryptor.finalize()

    plaintext = unpad(plaintext)

    with open(output_file, 'wb') as file:
        file.write(plaintext)

# Example usage
key = b'Sixteen byte key'
input_file = 'a.txt'
encrypted_file = 'encrypted.txt'
decrypted_file = 'decrypted.txt'

encrypt_file(input_file, encrypted_file, key)
decrypt_file(encrypted_file, decrypted_file, key)
```

```python
#6)program to implement steganography for hidding massage inside image file



from PIL import Image

def encode_message(image_path, message, output_path):
    img = Image.open(image_path)
    width, height = img.size

    message += '%%'
    binary_message = ''.join(format(ord(char), '08b') for char in message)

    if len(binary_message) > (width * height * 3):
        raise ValueError("Message is too long to be encoded in the image")

    index = 0
    for y in range(height):
        for x in range(width):
            pixel = list(img.getpixel((x, y)))
            for color in range(3):
                if index < len(binary_message):
                    pixel[color] = pixel[color] & ~1 | int(binary_message[index])
                    index += 1
            img.putpixel((x, y), tuple(pixel))

    img.save(output_path)

def decode_message(image_path):
    img = Image.open(image_path)
    width, height = img.size

    binary_message = ''

    for y in range(height):
        for x in range(width):
            pixel = img.getpixel((x, y))
            for color in pixel[:3]:
                binary_message += str(color & 1)

    end = binary_message.find('%%')
    binary_message = binary_message[:end]

    message = ''.join(chr(int(binary_message[i:i+8], 2)) for i in range(0, len(binary_message), 8))
    return message

# Example usage:

# Encode a message
encode_message('a.jpeg', 'Hello, this is a secret message!', 'output_image.png')

# Decode a message
decoded_message = decode_message('output_image.png')
print(decoded_message)
```

```python
#7)program to implement HMAC for signing message

import hmac
import hashlib

def sign_message(secret_key, message):
    hashed = hmac.new(secret_key.encode(), message.encode(), hashlib.sha256)
    return hashed.hexdigest()

# Example Usage
secret_key = "my_secret_key"
message = "Hello, World!"

signature = sign_message(secret_key, message)
print(f"Message: {message}")
print(f"Signature: {signature}")


```

```python

#8)program to implement  sending secure massage over ip network

from cryptography.fernet import Fernet
import socket

def generate_key():
    return Fernet.generate_key()

def encrypt_message(key, message):
    cipher_suite = Fernet(key)
    encrypted_message = cipher_suite.encrypt(message)
    return encrypted_message

def decrypt_message(key, encrypted_message):
    cipher_suite = Fernet(key)
    decrypted_message = cipher_suite.decrypt(encrypted_message)
    return decrypted_message

def send_secure_message(message, key, destination_ip, destination_port):
    encrypted_message = encrypt_message(key, message)
    with socket.socket(socket.AF_INET, socket.SOCK_STREAM) as s:
        s.connect((destination_ip, destination_port))
        s.sendall(encrypted_message)

def receive_secure_message(key, listen_ip, listen_port):
    with socket.socket(socket.AF_INET, socket.SOCK_STREAM) as s:
        s.bind((listen_ip, listen_port))
        s.listen()
        conn, addr = s.accept()
        with conn:
            encrypted_message = conn.recv(4096)
            decrypted_message = decrypt_message(key, encrypted_message)
            return decrypted_message

# Example usage:

key = generate_key()  # Generate a random key
message = b'Hello, this is a secure message.'

# Sender
send_secure_message(message, key, '127.0.0.1', 12345)

# Receiver
received_message = receive_secure_message(key, '127.0.0.1', 12345)
print(received_message.decode('utf-8'))
```

```python
import random

def is_prime(num):
    if num < 2:
        return False
    for i in range(2, int(num**0.5) + 1):
        if num % i == 0:
            return False
    return True

def find_primitive_root(p):
    # Find a primitive root modulo p
    for g in range(2, p):
        if all(pow(g, k, p) != 1 for k in range(1, p-1)):
            return g

def generate_keypair(p):
    g = find_primitive_root(p)
    private_key = random.randint(2, p-2)
    public_key = pow(g, private_key, p)
    return (p, g, public_key), (p, g, private_key)

def encrypt(public_key, plaintext):
    p, g, h = public_key
    y = random.randint(2, p-2)
    c1 = pow(g, y, p)
    s = pow(h, y, p)
    c2 = [s * ord(char) % p for char in plaintext]
    return c1, c2

def decrypt(private_key, c1, c2):
    p, g, x = private_key
    s = pow(c1, x, p)
    plaintext = [chr((c * pow(s, -1, p)) % p) for c in c2]
    return ''.join(plaintext)

if __name__ == '__main__':
    p = 23  # Example prime number
    public_key, private_key = generate_keypair(p)

    message = "HELLO"
    c1, c2 = encrypt(public_key, message)
    decrypted_message = decrypt(private_key, c1, c2)

    print(f"Original message: {message}")
    print(f"Encrypted message (c1, c2): ({c1}, {c2})")
    print(f"Decrypted message: {decrypted_message}")
```
